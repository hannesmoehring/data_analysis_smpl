```python
import pandas as pd
import os   
import numpy as np
import data_check as dc
```

```python
def load_texts(folder_path) -> pd.DataFrame:
    data = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            file_id = filename[:-4]
            with open(os.path.join(folder_path, filename), 'r', encoding='utf-8') as f:
                content = f.read().strip()
                content_desc = content.split('.')[0]
                content_types = content.split('.')[1]
            data.append({'id': file_id, 'content': content_desc, 'types' : content_types})

    df = pd.DataFrame(data)
    print(df.head())
    return df
```
